```python
import pandas as pd
```

```python
names = pd.Series(['Marks','Rita','Vicky','Justin','Jhosan','Michal'])
print(names)
```

```python
# Syntax-1 
print(names[0])
```

```python
print(names[5])
```

```python
# Syntax-2 
print(names[0:4])
```

```python
# Syntax-2 
print(names[0::2])
```

```python
# Syntax-3 
print(names[[0,4,5]])
```

```python
# Syntax-3 
print(names[[3,1,3]])
```

```python
names = pd.Series(['Marks','Rita','Vicky','Justin','Jhosan','Michal'],
                  index=["ID1","ID2","ID3","ID4","ID5","ID6"])
print(names,type(names))
```

```python
# Syntax-4
print(names[::2])
```
