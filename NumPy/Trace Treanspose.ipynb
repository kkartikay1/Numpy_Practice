```python
a=np.array([[1,2],[3,4]])
b=np.array([[11,12],[13,14]])
print("Matrix A\n",a)
print("Matrix B\n",b)
k=np.inner(a,b)
print("Inner Product Mtrix-A and Matix-B\n",k)
```
