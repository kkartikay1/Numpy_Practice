```python
import numpy as np
```

```python
a=np.array([10,20,30,40,50,60])
print("Oroganal Value =",a)
```

```python
# re-arange the element of ndarray
np.random.shuffle(a)
print("Shuffle ndarray object =",a)
```

```python
# re-arange the element of ndarray
np.random.shuffle(a)
print("Shuffle ndarray object =",a)
```

```python
for i in range(len(a)):
    np.random.shuffle(a)
    print("Shuffle ndarray object ={} at this Id={}".format(a,id(a)))
```

```python
# BY using Permutation re-arange the element of ndarray
arr=np.array([1,0,2,3,4,5])
x=np.random.permutation(arr)
print("Shuffle ndarray object =",arr)
print("Permuted ndarray Object =",x)
```

```python
# BY using Permutation re-arange the element of ndarray
arr=np.array(list("MISSISSIPPI"))
x=np.random.permutation(arr)
print("Shuffle ndarray object =",arr)
print("Permuted ndarray Object =",x)
```

```python
# BY using Permutation re-arange the element of ndarray
arr=np.array(list("MISSISSIPPI"))
for i in range(len(arr)):
    x=np.random.permutation(arr)
    # print("Shuffle ndarray object =",arr)
    print("Permuted ndarray Object =",x)
    print("".join(x),id(x))
    
```
