```python
import numpy as np
```

```python
# 1D- Array
a=np.array([10,2,13,4,-5,0,16,25,-16])
print(a,type(a))
```

```python
np.sort(a)
```

```python
print(a)
```

```python
np.sort(a)[::-1]
```

```python
# 2D- Array
a=np.array([10,2,13,4,-5,0,16,25,-16]).reshape(3,3)
print(a,type(a))
```

```python
np.sort(a)
```

```python
print(a)
```

```python
np.sort(a,axis=0) #here axiss=0 Represent Column
```

```python
np.sort(a,axis=1) #here axiss=1 Represent Row
```

```python
np.sort(a,axis=1)[::-1]
```

```python
a=np.array([10,20,3,4,11,15,16,7,8,-6,0,12]).reshape(2,3,2)
print(a)
```

```python
np.sort(a)
```

```python
np.sort(a,axis=0)
```

```python
print(a)
```

```python
np.sort(a,axis=1)
```

```python
a=np.array([10,20,3,4,11,15,16,7,8,-6,0,12]).reshape(3,4)
print(a)
```
