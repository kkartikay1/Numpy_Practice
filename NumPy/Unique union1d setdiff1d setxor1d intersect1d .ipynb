```python
import numpy as np
```

```python
a=np.array([10,20,30,40,20,10,60,40,20,30])
print(a,type(a))
```

```python
sa=np.unique(a)
print(sa,type(sa))
```

```python
a=np.array([10,20,30,40,20,10,60,40,20,30,20,10]).reshape(3,4)
print(a,type(a))
```

```python
sa=np.unique(a)
print(sa,type(sa))
```

```python
a=np.array([10,20,30,40,20,10,60,40,20,30,15])
print("Contend OF A =",a)
print("="*50)
sa,ind=np.unique(a,return_index=True)
print("Unique Elements=",sa)
print("Indices Of Unique Elements=",ind)
```

```python
a=np.array([10,20,30,40,20,10,60,40,20,30,20,10]).reshape(3,4)
print("Contend OF A =",a)
print("="*50)
sa,ind=np.unique(a,return_index=True)
print("Unique Elements=",sa)
print("Indices Of Unique Elements=",ind)
```

```python
a=np.array([10,20,30,40,20,10,60,40,20,30,15])
print("Contend OF A =",a)
print("="*50)
sa,ind=np.unique(a,return_inverse=True)
print("Unique Elements=",sa)
print("Indices Of Unique Elements=",ind)
```

```python
a=np.array([10,20,30,40,20,10,60,40,20,30,15])
print("Contend OF A =",a)
print("="*50)
sa,ind=np.unique(a,return_counts=True)
print("Unique Elements=",sa)
print("Indices Of Unique Elements=",ind)
```

```python
# Union
```

```python
a=np.array([10,20,30,40])
b=np.array([1,2,10,20])
c=np.union1d(a,b)
print(c)
```

```python
a=np.array([[10,20],[30,40]])
b=np.array([[1,2],[10,20]])
c=np.union1d(a,b)
print(c)
```

```python
# Intercsection
```

```python
a=np.array([10,20,30,40])
b=np.array([1,2,10,20])
c=np.intersect1d(a,b)
print(c)
```

```python
a=np.array([[10,20],[30,40]])
b=np.array([[1,2],[10,20]])
c=np.intersect1d(a,b)
print(c)
```

```python
# Finding set Differance
```

```python
a=np.array([10,20,30,40])
b=np.array([30,40,50,60])
c=np.setdiff1d(a,b)
print(c)
```

```python
a=np.array([10,20,30,40])
b=np.array([30,40,50,60])
c=np.setdiff1d(a,b,assume_unique=True)
print(c)
```

```python
a=np.array([10,20,30,40])
b=np.array([30,40,50,60])
c=np.setdiff1d(b,a)
print(c)
```

```python
# Finding Symmetric Difference
```

```python
a=np.array([10,20,30,40])
b=np.array([30,40,50,60])
c=np.setxor1d(a,b)
print(c)
```

```python
a=np.array([10,20,30,40])
b=np.array([30,40,50,60])
c=np.setxor1d(a,b,assume_unique=True)
print(c)
```

```python
a=np.array([10,20,30,40])
b=np.array([30,40,50,60])
c=np.setxor1d(b,a)
print(c)
```
