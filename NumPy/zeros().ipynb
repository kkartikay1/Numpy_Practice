```python
import numpy as np
```

```python
a=np.zeros(4)
print(a,type(a))
```

```python
a=np.zeros(4,dtype=int)
print(a,type(a))
```

```python
a=np.zeros(16,dtype=int).reshape(2,2,4)
print(a,type(a))
```

```python
a=np.zeros((2,2,5),dtype=int)
print(a,type(a))
print("Dimension :",a.ndim)
print("Shape :",a.shape)
print("Data type :",a.dtype)
print("Size :",a.size)
```
