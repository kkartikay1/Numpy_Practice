```python
lst=[10,20,30,40,50,60]
print(lst)
a=np.array(lst)
print(a,type(a))
print("-"*50)
print("Dimension :",a.ndim)
print("Shape :",a.shape)
print("Data type :",a.dtype)
```

```python
lst=[[10,20,30],[40,50,60],[80,90,100]]
print(lst)
a=np.array(lst)
print(a,type(a))
print("-"*50)
print("Dimension :",a.ndim)
print("Shape :",a.shape)
print("Data type :",a.dtype)
```
