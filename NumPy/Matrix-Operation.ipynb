```python
import numpy as np
```

```python
a=np.array([[11,20],[30,40]])
b=np.array([[50,60],[1,3]])
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix A \n ", b,"\n","-"*30)
```

```python
c=np.add(a,b)
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```

```python
d=a+c
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
print("Matrix D \n ", d,"\n","-"*30)
```

```python
a=np.array([[11,20],[30,40]])
b=np.array([[50,60],[1,3]])
c=np.subtract(a,b)
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```

```python
#Dot()
a=np.array([[11,20],[30,40]])
b=np.array([[1,2],[3,4]])
c=np.dot(a,b)
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```

```python
#matmul()
```

```python
a=np.array([[11,20],[30,40]])
b=np.array([[1,2],[3,4]])
c=np.matmul(a,b)
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```

```python
# Devide
```

```python
a=np.array([[11,20],[30,40]])
b=np.array([[1,2],[3,4]])
c=np.divide(a,b)  #Elementwise Division
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```

```python
a=np.array([[11,20],[30,40]])
b=np.array([[1,2],[3,4]])
c=a/b  #Elementwise Division by using / opertator insted to divide
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```

```python
a=np.array([[11,20],[30,40]])
b=np.array([[1,2],[3,4]])
c=np.floor_divide(a,b)  #Elementwise Division with Integer Quiotent
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```

```python
a=np.array([[11,20],[30,40]])
b=np.array([[1,2],[3,4]])
c=a//b  #Elementwise floor Division by using // opertator insted to divide
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```

```python
 # Modulo Devision
```

```python
a=np.array([[11,20],[30,40]])
b=np.array([[1,2],[3,4]])
c=np.mod(a,b)  #Elementwise Modulo Division 
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```

```python
a=np.array([[11,20],[30,40]])
b=np.array([[1,2],[3,4]])
c=a%b  #Elementwise Modulo Division  by using % opertator insted to divide
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```

```python
# Power Value
```

```python
a=np.array([[11,20],[30,40]])
b=np.array([[1,2],[3,4]])
c=np.power(a,b)  #Elementwise Exponention OR Power Value
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```

```python
a=np.array([[11,20],[30,40]])
b=np.array([[1,2],[3,4]])
c=a**b  #Elementwise Exponention OR Power Value by using ** operator insted by power operation
print("Matrix A \n ", a,"\n","-"*30)
print("Matrix B \n ", b,"\n","-"*30)
print("Matrix C \n ", c,"\n","-"*30)
```
