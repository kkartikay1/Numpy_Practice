```python
import numpy as np
```

```python
# Append
```

```python
a=np.array([10,20,30])
print(a,type(a))
```

```python
np.append(a,100)
print(a)
```

```python
a=np.append(a,100)
print(a,type(a))
```

```python
# Appending Multiple Array
```

```python
a=np.append(a,[100,200,300,400])
print(a,type(a))
```

```python
# 2D
a=np.array([[0,20],[30,40]])
print(a)
```

```python
a=np.append(a,[[50,60]],axis=0)
print(a)
```

```python
a=np.append(a,[[70],[80],[90]],axis=1)
print(a)
```

```python
arr=np.arange(1,13).reshape(2,6)
print("Orignal Array \n",arr)
```

```python
# Create Another Array Which is to Appebd column Wise
```

```python
col=np.arange(5,11).reshape(1,6)
print("Appended Array column wise \n",col)
```

```python
arr_col=np.append(arr,col,axis=0)
print("Array After Appending the values column wise")
print(arr_col)
```

```python
arr=np.arange(1,3).reshape(2,1)
print("Orignal Array \n",arr)
row=np.arange(3,9).reshape(2,3)
print("Appended Array Row wise \n",row)
arr_row=np.append(arr,row,axis=1)
print("Array After Appending the values Row wise")
print(arr_row)
```

```python
# Insert Operation
```

```python
a=np.array([10,20,30,40,50])
print(a,type(a))
```

```python
a=np.insert(a,1,15)
print(a,type(a))
```

```python
a=np.insert(a,3,[11,33,22,44])
print(a,type(a))
```

```python
a=np.array([10,20,30,40,50,60,70,80,90]).reshape(3,3)
print(a,type(a))
```

```python
a=np.insert(a,1,[11,22,33],axis=0)
print(a)
```

```python
a=np.insert(a,1,[33,44,55,66],axis=1)
print(a)
```

```python
# DELETE
```

```python
a=np.array([10,20,30,40,50,60,70,80,90])
print(a,type(a))
```

```python
a=np.delete(a,1)
print(a)
```

```python
a=np.delete(a,[0,-2,-1])
```

```python
print(a)
```

```python
a=np.arange(5*5).reshape(5,5)
print(a)
```

```python
a=np.delete(a,2,axis=0)
print(a)
```

```python
a=np.delete(a,[1,3],axis=0)
print(a)
```

```python
a=np.delete(a,[1,3],axis=1)
print(a)
```

```python
a=np.delete(a,[0,2],axis=1)
print(a)
```

```python
a=np.arange(5*5).reshape(5,5)
print(a)
print(a.shape)
print(a.shape[0])
print(a.shape[1])
```
