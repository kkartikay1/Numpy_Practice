```python
# MIN MAX
```

```python
# 1D Array 
```

```python
a=np.array([10,20,-3,45,-89,365,0,25,46,-5])
print(a)
```

```python
# 1D Array
a=np.array([10,20,-3,45,-89,365,0,25,46,-5])
print(a)
print("-"*50)
maxv=np.amax(a)
minv=np.amin(a)
print("Max Val = ",maxv)
print("Min Val = ",minv)
```

```python
 # 2D Array
a=np.array([[10,20,-3,45,-89],[365,0,25,46,-5],[564,-287,43,45,659],[385,0,25,596,-5000]])
print(a)
print("-"*20,"Max","-"*20)
colmax=np.amax(a,axis=0)
rowmax=np.amax(a,axis=1)
print("Max Val = ",colmax)
print("Min Val = ",rowmax)
print("-"*20,"Min","-"*20)
colmin=np.amin(a,axis=0)
rowmin=np.amin(a,axis=1)
print("Max Val = ",colmin)
print("Min Val = ",rowmin)
```

```python
# Findin Mean
```

```python
# 1D
a=np.array([20,20,30,40,50,10])
print("Content of a\n",a)
print("-"*50)
mv=np.mean(a)
print("Mean of a=",mv)
print("-"*50)
```

```python
 # 2D Array
a=np.array([[10,20,-3,45,-89],[365,0,25,46,-5],[564,-287,43,45,659],[385,0,25,596,-5000]])
print(a)
print("-"*20,"MEAN","-"*20)
mv=np.mean(a)
print("Mean of a=",mv)
print("-"*50)
```

```python
 # 2D Array
a=np.array([[10,20,-3,45,-89],[365,0,25,46,-5],[564,-287,43,45,659],[385,0,25,596,-5000]])
print(a)
print("-"*20,"MEAN","-"*20)
colmean=np.mean(a,axis=0)
rowmean=np.mean (a,axis=1)
print("Column Mean Val = ",colmean)
print("Row Mean Val = ",rowmean)
print("-"*50)
```

```python
# Variance
```

```python
 # 2D Array
a=np.array([[2,4],[1,3]])
print(a)
print("-"*20,"Varance","-"*20)
cVarance=np.var(a)
print("Varance Val = ",cVarance)
print("-"*50)
```

```python
 # 2D Array
a=np.array([[10,20,30],[2,15,95],[36,14,5]])
print(a)
print("-"*20,"Varance","-"*20)
colVarance=np.var(a,axis=0)
rowVarance=np.var(a,axis=1)
cVarance=np.var(a)
print("Varance Val = ",cVarance)
print("-"*50)
print("Column Varance Val = ",colVarance)
print("Row VaranceVal = ",rowVarance)
print("-"*50)
```

```python
# Stander Deviation
```

```python
 # 2D Array
a=np.array([[2,4],[1,3]])
print(a)
print("-"*20,"Stander Deviation","-"*20)
colstd=np.std(a,axis=0)
rowstd=np.std(a,axis=1)
cstd=np.std(a)
print("Stander Deviation Val = ",cstd)
print("-"*50)
print("Column Stander Deviation Val = ",colstd)
print("Row Stander Deviation Val = ",rowstd)
print("-"*50)
```
