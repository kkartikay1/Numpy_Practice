```python
import numpy as np
```

```python
arr1=np.array([10,30,50,60,80])
arr2=np.array([20,40,70,6,3])
newarr=np.add(arr1,arr2)
print(newarr)
```

```python
arr1=np.array([10,30,50,60,80])
arr2=np.array([20,40,70,6,3])
newarr=np.subtract(arr1,arr2)
print(newarr)
```

```python
arr1=np.array([10,30,50,60,80])
arr2=np.array([20,40,70,6,3])
newarr=np.multiply(arr1,arr2)
print(newarr)
```

```python
arr1=np.array([10,30,50,60,80])
arr2=np.array([20,40,70,6,3])
newarr=np.divide(arr1,arr2)
print(newarr)
```

```python
arr1=np.array([10,30,50,60,80,10])
arr2=np.array([20,40,70,6,3,0])
newarr=np.divide(arr1,arr2)
print(newarr)
```

```python
na=newarr.astype("str")
print(na)
na[-1]="Can't Devide"
print(na)
```

```python
arr1=np.array([10,30,50,60,80,60])
arr2=np.array([20,40,70,6,3,33])
newarr=np.power(arr1,arr2)
print(newarr)
```

```python
arr1=np.array([10,30,50,60,80,60])
arr2=np.array([20,40,70,6,3,33])
newarr=np.mod(arr1,arr2)
print(newarr)
```

```python
arr1=np.array([10,30,50,60,80,60])
arr2=np.array([20,40,70,6,3,33])
newarr=np.remainder(arr1,arr2)
print(newarr)
```
