```python
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([1, 2, 3])
newarr = np.add(arr1, arr2)
print(newarr)
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([1, 2, 3])
newarr = np.sum([arr1, arr2])
print(newarr)
```

```python
a=np.arange(3*3).reshape(3,3)
b=np.array([[1,0,0],[0,1,0],[0,0,1]])
print("Matrix A \n",a)
print("Matrix B \n",b)
```

```python
c=np.add(a,b)
print("Matrix c \n",c)
```

```python
e=np.sum([a,b])
print("Matrix D \n",e)
```

```python
d=np.sum((a,b),axis=1)
print("Matrix D \n",d)
```

```python
d=np.sum(np.sum((a,b),axis=1),axis=1)
print("Matrix D \n",d)
```

```python
d=np.sum(np.sum(np.sum((a,b),axis=1),axis=1),axis=0)
print("Matrix D \n",d)
```

```python
arr1 = np.array([1, 2, 3,4])
newarr = np.cumsum(arr1)
print(newarr)
```

```python
arr1 = np.array([1, 2, 3,4])
newarr = np.cumsum(arr1,n=3)
print(newarr)
```
