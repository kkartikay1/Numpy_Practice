```python
# Differances B/w flatten() and Ravel()
a=np.arange(12).reshape(4,3)
print("Orignal Array \n",a)
print("-"*50)
print("Flatten Array")
flt=a.flatten()
flt[0]=10
print(flt)
print("-"*50)
print("Orignal Array \n",a)
print("-"*50)
print("Ravel Array")
rav=a.ravel()
rav[0]=20
print(rav)
print("-"*50)
print("Orignal Array \n",a)
```
