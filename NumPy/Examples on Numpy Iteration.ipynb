```python
import numpy as np
```

```python
a=np.array([10,20,30,40,50,60])
print(a)
print("Using for Loop")
for val in a:
    print(val)
print("Using while Loop")
i=0
while(i<len(a)):
    print(a[i])
    i=i+1
print("Using while Loop")
i=0
while(i<len(a)):
    print(a[i],end="-->")
    i=i+1
```

```python
a=np.arange(16).reshape(4,4)
print("Contant of a\n",a)
print("Dimension =",a.ndim)
print("Shape =",a.shape)
```

```python
for val in a:
    for inr in val:
        print(inr)
```

```python
a=np.arange(16).reshape(2,4,2)
print("Contant of a\n",a)
print("Dimension =",a.ndim)
print("Shape =",a.shape)
```

```python
for val in a:
    for inr in val:
        for dinr in inr:
            print(dinr)
```

```python
# Iteration Using NumPy Model
```

```python
a=np.arange(5)
print(a,type(a))
```

```python
for val in np.nditer(a):
    print(val)
```

```python
for index,value in np.ndenumerate(a):
    print(index,"-->",value)
```
