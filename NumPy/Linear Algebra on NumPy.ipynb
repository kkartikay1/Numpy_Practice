```python
import numpy as np
```

```python
# Consider the Following Liner Equation
# 2x+3y=6
# 6x+4y=3
# find x and Y values
```

```python
a=np.array([[2,3],[6,4]])
b=np.array([6,3])
print(a,type(a))
print(b,type(b))
```

```python
# SOlve
sol=np.linalg.solve(a,b)
print(sol)
```

```python
a=np.array([[2,3],[6,4]])
b=np.array([6,2])
print(a,type(a))
print(b,type(b))
# SOlve
sol=np.linalg.solve(a,b)
print(sol)
```

```python
# Consider
# x+y+z=6
# 2x+3y+9z=16
# 2x-4y-9z=20
a=np.array([[1,1,1],[2,3,9],[2,-4,-9]])
b=np.array([6,16,20])
print(a,"\n",b)
```

```python
# Solve
```

```python
sol=np.linalg.solve(a,b)
print(sol)
```

```python
# Consider
# x+y+z=6
# 2x+3y+9z=16
# 2x-4y-9z=20
a=np.array([[1,1,1],[2,3,9],[2,-4,-9]])
b=np.array([6,16,20])
print(a,"\n",b)
 # Solve
sol=np.linalg.solve(a,b)
print("Value of x=",sol[0])
print("Value of y=",sol[1])
print("Value of z=",sol[2])
```

```python
# Find Det of Given Matrix
a=np.array([[1,2],[3,4]])
print("Given Matrix\n",a)
```

```python
# solve
deta=np.linalg.det(a)
print("Det of a =",deta)
```
