```python
import numpy as np
```

```python
a=np.arange(10)
print(a,type(a))
```

```python
a
```

```python
print("Dimension :",a.ndim)
print("Shape :",a.shape)
print("Size :",a.size)
print("Data type :",a.dtype)
```

```python
a.shape=(5,2)
print(a,type(a))
print("Dimension :",a.ndim)
print("Shape :",a.shape)
print("Size :",a.size)
print("Data type :",a.dtype)
```

```python
a=a.reshape(2,5)
print(a,type(a))
print("Dimension :",a.ndim)
print("Shape :",a.shape)
print("Size :",a.size)
print("Data type :",a.dtype)
```
