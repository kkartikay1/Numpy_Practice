```python
a=np.array([10,20,30,40])
b=np.array([100,200,300,400])
c=np.hstack((a,b))
print(c,type(c))
```

```python
a=np.array([[10,20,30],[40,50,60],[70,80,90]])
b=np.array([[100,200],[300,400],[500,600]])
print("Matrix -A")
print(a,type(a),a.shape)
print("="*50)
print("Matrix -B")
print(b,type(b),b.shape)
print("="*50)
c=np.hstack((a,b))
print("Matrix -c")
print(c,type(c),c.shape)
```
